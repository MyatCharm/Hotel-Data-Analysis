```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df2018 = pd.read_csv("https://raw.githubusercontent.com/MyatCharm/Hotel-Data-Analysis/main/hotel_revenue_2018.csv")
df2018
```

```python
df2019 = pd.read_csv("https://raw.githubusercontent.com/MyatCharm/Hotel-Data-Analysis/main/hotel_revenue_2019.csv")
df2019
```

```python
df2020 = pd.read_csv("https://raw.githubusercontent.com/MyatCharm/Hotel-Data-Analysis/main/hotel_revenue_2020.csv")
df2020
```

```python
df_hotel = pd.concat([df2018, df2019, df2020])
df_hotel
```

```python
df_hotel.columns
```

```python
df_market_segment = pd.read_csv("https://raw.githubusercontent.com/MyatCharm/Hotel-Data-Analysis/main/hotel_revenue_market_segment.csv")
df_market_segment
```

```python
df_meal_cost = pd.read_csv("https://raw.githubusercontent.com/MyatCharm/Hotel-Data-Analysis/main/hotel_revenue_meal_cost.csv")
df_meal_cost
```

```python
df_hotel1 = pd.merge(df_hotel, df_market_segment, how='left',on=['market_segment','market_segment'])
df_hotel = pd.merge(df_hotel1, df_meal_cost, how='left',on=['meal','meal'])
df_hotel
```

```python
df_hotel['Revenue'] = (((df_hotel['stays_in_weekend_nights']+df_hotel['stays_in_week_nights'])*df_hotel['adr'])*df_hotel['Discount'])-df_hotel['Cost']
df_hotel
```

```python
df_revenue = df_hotel.groupby(['hotel','arrival_date_year'])['Revenue'].sum().reset_index(name ='Revenue')
df_revenue
```

```python
df_revenue ['in ($)millions'] = '$ ' + (df_revenue['Revenue'].astype(float)/1000000).round(2).astype(str) + ' M'
df_revenue
```

```python
sns.lineplot(data=df_hotel, x="arrival_date_month", y="Revenue",hue="arrival_date_year", style="hotel")
plt.show()
```

```python
df_parking = df_hotel.groupby(['hotel','arrival_date_year'])['required_car_parking_spaces'].sum().reset_index(name ='Parking Space Need')
df_parking
```
